# Quantum Behavioral Analysis Demo

This notebook shows how to use quantum_behavioral_analysis.py on both agent/user log data and task update streams. Requires qiskit or falls back to classical kernel.


```python
from quantum_behavioral_analysis import BehavioralAnalyzer
import numpy as np
import random
```

## Example: Encode interaction logs (dummy data)


```python
np.random.seed(42)
user_events = [np.random.rand(random.randint(3,6)) for _ in range(5)]
analyzer = BehavioralAnalyzer(use_quantum=True)
sim = analyzer.analyze(user_events)
analyzer.summary(sim)
```

## Example: Task update streams (dummy)


```python
task_update_sequences = [np.random.rand(random.randint(4,7)) for _ in range(4)]
sim2 = analyzer.analyze(task_update_sequences)
analyzer.summary(sim2)
```
